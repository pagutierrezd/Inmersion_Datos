```python
# Consultamos nuevamente la información del dataset
inmuebles.info()
```

```python
# Antes de realizar la conversión a un objeto númerico es necesario realizar un tratamiento previo, en este caso se eliminan los puntos
# de los precios para que no cause problemas en el futuro y se generen confusiones con las separaciones de millares y el punto decimal
inmuebles['Precio'] = inmuebles['Precio'].str.replace('.','',regex=True)

# Consultando el dataset para observar los cambios
inmuebles[['Precio','Barrio']] # El doble corchete es para indicar que el resultado sea un dataframe
```

```python
# Hasta este momento el atributo "Valor" sigue siendo objeto
inmuebles.info()
```

```python
# Se crea una nueva columna en el dataset que corresponde al precio en millones de los inmuebles, además se realiza la conversión
# a tipo flotante para trabajar con valores númericos
inmuebles['Precio_Millon'] = inmuebles.Precio.astype('float')/1000000

# Con la información es posible notar las características de la nueva colummna que hemos creado
inmuebles.info()
```

```python
# Precio del m2 por barrio pero teniendo en cuenta sólo el precio y área del propio inmueble y no los de todo el barrio
inmuebles['Valor_m2_Millon'] = inmuebles['Precio_Millon']/inmuebles['Area'] 
inmuebles.head(3)
```

```python
# Agrupando por barrio y obteniendo el promedio para todas las columnas númericas, en este caso se ha obtenido el promedio de medias para todo el barrio
# En la columna valor_m2_millon se aprecia el precio por m2 de cada barrio
inmuebles.groupby('Barrio').mean()
```

```python
# Agrupamos por barrio y sumamos los valores
datos_barrio = inmuebles.groupby('Barrio').sum()

datos_barrio
```

```python
# Calculando el precio del m2 por barrio pero teniendo en cuenta la suma total de precios y área de todo el barrio
# En la columna valor_m2_barrio se aprecia el precio por m2 de cada barrio
datos_barrio['Valor_m2_Barrio'] = datos_barrio['Precio_Millon']/datos_barrio['Area']
datos_barrio
```

*Sí se desea obtener la perspectiva general del precio por metro cuadrado para cada barrio es recomendable obtenerlo sumando primero el área total del barrio y el precio de los inmuebles, para posteriormente obtener el precio por m2, tal como se especifica en la opción 2.*

*Si lo que nos interesa es el precio del m2 para cada inmueble es recomendable la opción 1 pues trabaja con los registros a nivel individual*

```python
# Creando un diccionario con el precio por m2 de cada barrio
# {clave:valor} = {barrio:precio}
m2_barrio = dict(datos_barrio['Valor_m2_Barrio'])
m2_barrio
```

```python
# Mapeando los precios por m2 de cada barrio a cada inmueble en el dataset original
inmuebles['Valor_m2_Barrio'] = inmuebles['Barrio']
inmuebles['Valor_m2_Barrio'] = inmuebles['Valor_m2_Barrio'].map(m2_barrio)
inmuebles.head()
```

```python
# El top 10 de barrios con más inmuebles a la venta
top_barrios = inmuebles['Barrio'].value_counts()[:10].index
top_barrios
```

```python
# Retomando el PandasSeries y reseteando el índice para que Barrio deje de ser índice y se convierta en una columna 
datos_barrio.reset_index(inplace=True)
datos_barrio
```

```python
# El top 10 de barrios 
datos_barrio.query('Barrio in @top_barrios') # query('column in @variableReferencia')
```

```python
# Creando un gráfico para visualizar los datos de nuestro nuevo dataframe
plt.figure(figsize=(10,8))

# Se compara el barrio y el precio x m2 del top 10 barrios con mayor cantidad de inmuebles a la venta
ax = sns.barplot(x="Barrio", y="Valor_m2_Barrio", data = datos_barrio.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
```

```python
# Creando un boxplot para visualizar la distribución de los precios por m2 de los inmuebles según su barrio
#   - Las líneas extremas muestran justamente los valores extremos del precio
#   - La línea inicial de la caja índica el 25%, la segunda índica el 50% o mediana y la tercera el 75%
#   - Los puntos fuera del diagrama se consideran outlier o valores atípicos fuera del comportamiento "normal" de los demás valores
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Valor_m2_Millon", data = inmuebles.query('Barrio in @top_barrios'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Agregando una condición para mejorar el gráfico y eliminar los outliers
# Colocando límite al valor por m2
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Valor_m2_Millon", data = inmuebles.query('Barrio in @top_barrios & Valor_m2_Millon < 15'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Boxplot con la distribución del área de los inmuebles por barrio
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Area", data = inmuebles.query('Barrio in @top_barrios & Area < 500'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```

```python
# Boxplot con la distribución del precio de los inmuebles por barrio
plt.figure(figsize=(10,8))
ax = sns.boxplot(x="Barrio", y="Precio_Millon", data = inmuebles.query('Barrio in @top_barrios & Precio_Millon < 2000'))
ax.tick_params(axis='x', rotation=45)
plt.show()
```
